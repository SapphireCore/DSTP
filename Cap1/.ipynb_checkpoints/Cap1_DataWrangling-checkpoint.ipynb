## Background
The file compiled contained complete public loan data for all loans issued by Lending Club from 2007 to 2015, including the current loan status (Current, Late, Fully Paid, etc.). The file has about 890 thousand observations and 75 variables.

The target variable is **loan_status**.

## Load in Dataset
In this step we will load in the stored csv file and examine the first and last a few rows.  
A lot of missing values (NaN) shown on a number of variables. These need to be investigated to see how much information is included in each of these variables.

```python
import pandas as pd
loan = pd.read_csv("../LendingClubLoan/loan.csv", low_memory=False)

# Ensure all columns' info will be shown.
# The data description indicated that there are 74 columns.
pd.set_option('display.max_columns', 100)

# Check whether dataset is loaded successfully.
loan.head(10)
#loan.tail(10)
```

## Data Wrangling
### Descripency: dataset vs. data dictionary
The loaded dataframe has 887379 rows and 74 columns. However, the data dictionary provided has 78 variables. Through careful examniation, the following variables are found not included: 

* fico_range_high  
* fico_range_low  
* is_inc_v  
* last_fico_range_high  
* last_fico_range_low  

Also, **verification_status** has no definition. Notice that *verification_status_joint* has only 511 non-null values and the categories are identical to *verification_status*, *verification_status_join* is going to be dropped.

```python
loan.describe(include = "all")
```

```python
loan.info()
```

```python
loan.drop(["verification_status_joint"], axis = 1, inplace = True)
```

### Check missing values
Based on the nun-null value counts, we decide to drop variables that have less than 5% information (i.e. 95%+ null).

```python
exclude_var = (loan.isnull().sum(axis = 0) / loan.shape[0]).sort_values(ascending = False)
exclude_var = exclude_var[exclude_var >= 0.95]
loan.drop(exclude_var.index, axis = 1, inplace = True)
```

```python
loan.info()
```

### Drop other variables that are irrelevant to the analysis
The following variables, after carefully examination, are considered irrelevant to this analysis and thus will be dropped from dataset.

1. *id*: this serves as an index column and all values are unique. No information is provided related to *loan_status*.
2. *member_id*: similar to *id*, unique value each row, no useful information.
3. *url*: the url of each loan's page. No information may be parsed from this string.
4. *desc*: loan description. Only 15% row has this info. Considering that there is a variable *purpose* that serves the loan purpose grouping, it would be safe to drop *desc*.
5. *title*: loan title. The information could be represented by *purpose* as well and may be dropped.
6. *zipcode*: there is another variable *addr_state* that has better geographical information. May be dropped.
7. *policy_code*: = 1 for all rows. No use.
8. *pymnt_plan*: mostly n and only 10 y. Almost no information.
9. *application_type*: mostly INDIVIDUAL, only ~500 JOINT. Almost no information.

```python
toDrop = ['id', 'member_id', 'url','desc','title', 'zip_code','policy_code', 'pymnt_plan', 'application_type']
loan.drop(toDrop, axis = 1, inplace = True)
```

```python
loan.info()
```

### Fix data types

Convert date string to datetime type.

```python
loan['issue_d'] = pd.to_datetime(loan['issue_d'])
loan['last_pymnt_d'] = pd.to_datetime(loan['last_pymnt_d'])
loan['next_pymnt_d'] = pd.to_datetime(loan['next_pymnt_d'])
loan['last_credit_pull_d'] = pd.to_datetime(loan['last_credit_pull_d'])
```

Strip term as numbers instead of strings.
Convert interst rate to 0-1 range by dividing by 100.

```python
loan['term'] = loan['term'].str.split(' ').str[1]
loan['int_rate'] = loan['int_rate'] / 100.0
```
