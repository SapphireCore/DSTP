### Categorical Features

```python
# Construct a list of categorical features
cat_features = loan.select_dtypes(include=["object"]).columns

# Show each categorical variable's distributions against the loan_status_simple
for cat_feature in cat_features:
    print(loan.groupby(cat_feature).size().sort_values())
```

The following features should be dropped:  
1. grade and sub_grade. Because they are risk indicators and at the application these features would not be available.
2. emp_title. This is a very messy text column. Without careful feature engineering to categorize the titles into a small number of groups it won't be as meaningful. At this stage we could simply drop it.
3. Date features. Except the year information the nature of the data does not form time series. Therefore these features could simply be dropped.

The following features need to be engineered:  
1. emp_length. Should extract the year value, recode "< 1 year" as 0 and "10+ year" as 10, then fill the n/a as the mean. (Object to Integer)
2. home_ownership. Need to consolidate ANY, NONE and OTHER as OTHER to reduce the category number.
3. earliest_cr_line. Combine this feature and issue_d to calculate length of credit history in years (cr_hist_yr), calculated as cr_hist_yr = issue_d.year() - earliest_cr_line.year().

```python
# Drop grade, sub_grade and emp_title.

loan.drop(["grade", "sub_grade", "emp_title"], axis = 1, inplace = True)
```

```python
# Create emp_length_yr from emp_length.
import numpy as np

emp_length_dict = {
    "9 years": 9,
    "6 years": 6,
    "8 years": 8,
    "7 years": 7,
    "n/a": np.nan,
    "4 years": 4,
    "5 years": 5,
    "1 year" : 1,
    "3 years": 3,
    "< 1 year": 0,
    "2 years": 2,
    "10+ years": 10
}

loan["emp_length_yr"] = loan["emp_length"].map(emp_length_dict)
loan.drop("emp_length", axis = 1, inplace = True)
```

```python
# Consolidate levels for home_ownership

home_own_dict = {
    "ANY": "OTHER",
    "NONE": "OTHER",
    "OTHER": "OTHER",
    "OWN": "OWN",
    "RENT": "RENT",
    "MORTGAGE": "MORTGAGE"
}

loan["home_owner_s"] = loan["home_ownership"].map(home_own_dict)
loan.drop("home_ownership", axis = 1, inplace = True)
```

```python
# Calculate length of credit history in years (cr_hist_yr)

loan['cr_hist_yr'] = loan.issue_d.dt.year - loan.earliest_cr_line.str.split('-').str[1].astype("float")
loan.drop("earliest_cr_line", axis = 1, inplace = True)
```

```python
# Drop date features
exclude = ["issue_d", "last_pymnt_d", "next_pymnt_d", "last_credit_pull_d"]
loan.drop(exclude, axis = 1, inplace = True)
```
