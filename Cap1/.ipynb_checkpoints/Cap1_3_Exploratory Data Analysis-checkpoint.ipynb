### Numerical Features

Although most of the features have many 0's, the ranges seem to be reasonable. Through standard scaling the effects of different ranges of the features would be eliminated.

```python
import matplotlib.pyplot as plt

loan.hist(bins=50, figsize=(20, 15))
plt.show()
```

Check these variables and see whether their maxima are within reasonable range: dti, mth_since_last_delinq, pub_rec, recoveries, revol_bal, tot_coll_amt, tot_cur_bal, total_rec_late_fee, total_rev_hi_lim.

```python
ck_lst = ['dti', 'mths_since_last_delinq', 'pub_rec', 'recoveries', 'revol_bal', 'tot_coll_amt', 'tot_cur_bal', 'total_rec_late_fee', 'total_rev_hi_lim']

for var in ck_lst:
    print(var)
    print(loan[var].max())
```

dti and total_rev_hi_lim show abnormal patterns. The maximum values are 9999.0 and 9999999.0. dti is debt to income (excluding mortgage and large loans), and total_rev_hi_lim is total revolving high credit/credit limit. The two maximum values seem to be fillers for max-out ceilings and should be further investigated.

```python
# Sort dti in descending order.
loan.dti.sort_values(ascending=False).head()
```

```python
# Sort total_rev_hi_lim in descending order.
loan.total_rev_hi_lim.sort_values(ascending=False).head()
```

It seems only a few observations have the max-out coding. Using the maximum values (excluding 9999 or 9999999) to replace these max-out values should be reasonable for standard scaling.

```python
loan.dti.replace(9999.0, 1092.52, inplace=True)
loan.total_rev_hi_lim.replace(9999999.0, 2013133.0, inplace=True)
print(loan.dti.max())
print(loan.total_rev_hi_lim.max())
```

Now we need to fill those missing values. A common practice is to use the mean of each feature to fill.

```python
# Impute missing values of numerical features using column means
num_features = loan.select_dtypes(include=["float64"]).columns

for num_feature in num_features: 
    loan[num_feature] = loan[num_feature].fillna(loan[num_feature].mean())
```

After filling the missing values, a quick exam of these numerical features vs. the target using box plots helps us find whether any of them may be indicative.

There seem to be a few features that are indicative, such as recoveries and collection_recovery_fee. However, from the nature of loan, default loan will surely have recovries / collection fee but the good loan won't. It is clearly data leakage here. Other than that, no other feature seems to be indicative.

```python
# Check each numerical feature vs. loan_status_simple
for num_feature in num_features: 
    loan.boxplot(column = num_feature, by = 'loan_status_simple')
```

### Categorical Features

```python
# Construct a list of categorical features
cat_features = loan.select_dtypes(include=["object"]).columns

# Show each categorical variable's distributions against the loan_status_simple
for cat_feature in cat_features:
    print(loan.groupby(cat_feature).size().sort_values())
```

The following features should be dropped:  
1. grade and sub_grade. Because they are risk indicators and at the application these features would not be available.
2. emp_title. This is a very messy text column. Without careful feature engineering to categorize the titles into a small number of groups it won't be as meaningful. At this stage we could simply drop it.
3. Date features. Except the year information the nature of the data does not form time series. Therefore these features could simply be dropped.

The following features need to be engineered:  
1. emp_length. Should extract the year value, recode "< 1 year" as 0 and "10+ year" as 10, then fill the n/a as the mean. (Object to Integer)
2. home_ownership. Need to consolidate ANY, NONE and OTHER as OTHER to reduce the category number.
3. earliest_cr_line. Combine this feature and issue_d to calculate length of credit history in years (cr_hist_yr), calculated as cr_hist_yr = issue_d.year() - earliest_cr_line.year().

```python
# Drop grade, sub_grade and emp_title.

loan.drop(["grade", "sub_grade", "emp_title"], axis = 1, inplace = True)
```

```python
# Create emp_length_yr from emp_length.
import numpy as np

emp_length_dict = {
    "9 years": 9,
    "6 years": 6,
    "8 years": 8,
    "7 years": 7,
    "n/a": np.nan,
    "4 years": 4,
    "5 years": 5,
    "1 year" : 1,
    "3 years": 3,
    "< 1 year": 0,
    "2 years": 2,
    "10+ years": 10
}

loan["emp_length_yr"] = loan["emp_length"].map(emp_length_dict)
loan.drop("emp_length", axis = 1, inplace = True)
```

```python
# Consolidate levels for home_ownership

home_own_dict = {
    "ANY": "OTHER",
    "NONE": "OTHER",
    "OTHER": "OTHER",
    "OWN": "OWN",
    "RENT": "RENT",
    "MORTGAGE": "MORTGAGE"
}

loan["home_owner_s"] = loan["home_ownership"].map(home_own_dict)
loan.drop("home_ownership", axis = 1, inplace = True)
```

```python
# Calculate length of credit history in years (cr_hist_yr)

loan['cr_hist_yr'] = loan.issue_d.dt.year - loan.earliest_cr_line.str.split('-').str[1].astype("float")
loan.drop("earliest_cr_line", axis = 1, inplace = True)
```

```python
# Drop date features
exclude = ["issue_d", "last_pymnt_d", "next_pymnt_d", "last_credit_pull_d"]
loan.drop(exclude, axis = 1, inplace = True)
```

Similar to numerical features, we need to fill the missing values using "MISSING" coding for later dummy coding.

```python
# Construct a list of categorical features
cat_features = loan.select_dtypes(include=["object"]).columns.drop("loan_status_simple")

# Fill missing values of categorical features using MISSING as a category
for cat_feature in cat_features: 
    loan[cat_feature] = loan[cat_feature].fillna("MISSING")
```

Now check whether any categorical feature is indicative to predict target. It seems nothing is worth noting here.

```python
# Check each categorical feature vs. target
for cat_feature in cat_features:
    print(pd.crosstab(index=loan[cat_feature], columns=loan["loan_status_simple"]))
    print("****************")
```
