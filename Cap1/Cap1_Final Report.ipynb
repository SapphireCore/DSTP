### BACKGROUND
Lending Club (NYSE: LC) is a US peer-to-peer lending company that registered its offering as securities with the Securities and Exchange Commission to offer loan trading on a secondary market. It is by far the world's largest peer-to-peer lending platform. The reported revenue in 2016 is US$ 501 million.

Lending Club enables borrowers to apply for loans between US\$1,000 and US\$40,000, usually in three-year period. Its risk assessment model categorizes borrowers by assigning them a loan grade (and a sub grade) based on their credit history. Investors can search the loan listings on Lending Club platform to select loans they intend to invest, based on the information provided by Lending Club such as amount of loan, loan grade, purpose, etc. The investment profit gained by investors is the loan interests.

### BUSINESS PROBLEMS
The Risk Analysis Team of Lending Club company would like to build a predictive model for loan default. Particularly, any late payment more than 30 days would be considered as risky and potentially leads to default in the future. By constantly monitoring the current loan status and variables changing over time, Lending Club would have a better control on the loan default risk.

### DATA DESCRIPTION
The dataset was obtained from Kaggle, which is a combined dataset of loans issued between 2007 and 2015 downloaded from Lending Club website. The dataset includes the current loan status (Current, Late, Fully Paid, etc.) and latest payment information as well as borrowers' past credit history. Totally there are more than 800,000 rows and 74 columns. A summary data dictionary is provided describing the meanings and possible values of each variable.

### DATA WRANGLING
First, the data integrity was examined and some discrepancy was found. The variables without descriptions were removed from dataset.

Secondly, based on the nun-null value counts, we decide to drop variables that have less than 5% information (i.e. 95%+ null).

Then, after carefully examination, the following variables are considered irrelevant to this analysis and thus will be dropped from dataset.

1. *id*: this serves as an index column and all values are unique. No information is provided related to *loan_status*.
2. *member_id*: similar to *id*, unique value each row, no useful information.
3. *url*: the url of each loan's page. No information may be parsed from this string.
4. *desc*: loan description. Only 15% row has this info. Considering that there is a variable *purpose* that serves the loan purpose grouping, it would be safe to drop *desc*.
5. *title*: loan title. The information could be represented by *purpose* as well and may be dropped.
6. *zipcode*: there is another variable *addr_state* that has better geographical information. Considering ththe zipcode is partially masked, it is dropped.
7. *policy_code*: = 1 for all rows. No use.
8. *pymnt_plan*: mostly n and only 10 y. Almost no information.
9. *application_type*: mostly INDIVIDUAL, only ~500 JOINT. Almost no information.

Some variables have incorrect data types and were corrected. Date strings were converted to datetime type. Strip term as numbers instead of strings. Convert interst rate to 0-1 range by dividing by 100.

### PROCESS CATEGORICAL VARIABLES
The category distributions of each categorical variable were examined using frequency tables. From the output, we see that the following variables were dropped: grade and sub_grade, because they are risk indicators and at the application these features would not be available; emp_title, which is a very messy text column, and without careful feature engineering to categorize the titles into a small number of groups it won't be as meaningful; and all date variables, because except the year information the nature of the data does not form time series. 

And, the following variables were engineered: emp_length, by extracting the year value, recode "< 1 year" as 0 and "10+ year" as 10, then fill the n/a as the mean; home_ownership, by consolidating ANY, NONE and OTHER as OTHER to reduce the category number; and earliest_cr_line, by combining this variable and issue_d to calculate length of credit history in years (cr_hist_yr), calculated as cr_hist_yr = issue_d.year() - earliest_cr_line.year(). The missing values occurred in the two newly engineered numerical variables were filled using column means.

We filled the missing values in each of the rest categorical variable using "MISSING" coding. It will be dropped in a later stage when doing dummy coding.

### TRAINING / TEST SET PREPARATION
We have seen in the previous section that the target (loan_status_simple) is not balaced and including a level Issued that needs to be removed from the target. We first defined custom functions for standardization and dummy coding, and applied these functions on the imputed dataset. Then, we separated the dataset by loan status into three datasets. The dataset with status = Issued was removed from the modeling, and the dataset with status = Good was undersampled to 10% of original size which yielded 81149 rows. The undersampled dataset was appended to the dataset with status = Bad to form the train_test_set. The train_test_set was then splitted into X_train, y_train, X_test and y_test for further modeling. The split proportion used was 30%.

### FEATURE SELECTION
We used Extra Trees Classifier to build the first model with all features, and outputed the feature importance from high to low. Based on some experiments of cutoff values, it seems cutoff = 0.05 could yield an optimal subset of features without impacting the predicting accuracy. The selected "optimal feature list" was then used for further model tuning and performance evaluation.

### MODEL TUNING & SELECTION
Based on the general performance and training speed found in literature, a few model candidates were identified: logistic regression, random forest, support vector machine (linear), and naive Bayesian. A number of hyperparameters with possible value spaces were included in the grid to search for the best hyperparameter combinations for each model. The model with the best performance measured by 10-fold crossvalidation mean accuracy among the four was then picked as the overall best model for prediction.

The overall best model was random forest with gini index as the split criterion, maximum tree depth of 10, max feature as sqrt, minimum 5 samples in each leaf, minimum 3 samples to split an internal node, and the number of trees in the forest of 9. This model outputs a 10-fold cv accuracy of 82.5% for training set. 

### PREDICTION RESULT
The above mentioned best model was then used to predict the test set and evaluated against the actual value as the final model evaluation. The ROC curve was plotted and related area under curve were calculated.

The prediction accuracy on the test set was 84.2%, which was about the same as the 10-fold cv accuracy of training set. The precision and recall were 94.0% and 69.6%, respectively, which indicated that about 94% of the default loans were identified correctly.
